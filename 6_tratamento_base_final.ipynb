```python


# Agrupando pelas colunas e contando o número de ocorrências
contagem = merge_exploded_cerveja.groupby(['estilos_cervejas_brejas', 'pratos', 'category']).size().reset_index(name='count')

# Filtrando para mostrar apenas combinações que aparecem mais de uma vez
contagem_filtrada = contagem[contagem['count'] > 1]

# Ordenando do maior para o menor pela coluna 'count'
contagem_ordenada = contagem_filtrada.sort_values(by='count', ascending=False)

# Exibindo o resultado
print(contagem_ordenada)
```

```python


# Agrupando por estilos_cervejas_brejas e category e somando as contagens
contagem_final = contagem_filtrada.groupby(['estilos_cervejas_brejas', 'category'])['count'].sum().reset_index()

# Ordenando do maior para o menor pela coluna 'count'
contagem_ordenada = contagem_final.sort_values(by='count', ascending=False)

# Exibindo o resultado
print(contagem_ordenada)
```
