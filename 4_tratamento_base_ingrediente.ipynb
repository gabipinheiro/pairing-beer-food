```python
import pandas as pd
from collections import Counter
from sklearn.preprocessing import MinMaxScaler

```

```python
# carregando o dataset
ingredientes = pd.read_csv('./datasets/preprocessed/INGREDIENTES_preprocessed.csv')
```

```python
ingredientes_chocolate = ingredientes[ingredientes['ingrediente'] == 'chocolate']
```

```python
# função para expandir os sabores e pesos

def expand_peso_sabores(peso_sabores):
    # avaliar o Counter para obter o dicionário de pesos
    return eval(peso_sabores)
    
```

```python
# aplicar a função para expandir a coluna 'peso_sabores'
ingredientes_expandido = ingredientes['peso_sabores'].apply(expand_peso_sabores)
```

```python
# criando um DataFrame a partir dos dicionários de pesos de sabores
ingredientes_sabores_df = pd.DataFrame(ingredientes_expandido.tolist())
```

```python
# preenchendo valores NaN com 0 (ingredientes que não têm aquele sabor)
ingredientes_sabores_df = ingredientes_sabores_df.fillna(0)

```

```python
# anexando as colunas 'ingrediente' e 'categoria' ao novo DataFrame
ingredientes_sabores_df = pd.concat([ingredientes[['ingrediente', 'categoria']], ingredientes_sabores_df], axis=1)

```

```python
ingredientes_sabores_df[['ingrediente','categoria','sour', 'bitter', 'sweet', 'fruity', 'grassy','roast', 'caramel', 'fermented','powdery','dry']]
```

```python
# salvando a base tratada em csv
ingredientes_sabores_df.to_csv('./datasets/preprocessed/INGREDIENTES_processed.csv', index=False)
```

```python
todos_sabores = [col for col in ingredientes_sabores_df.columns if col not in ['ingrediente', 'categoria']]
```

```python
# normalizando os valores para todos se encaixarem entre 0 e 1, respeitando a proporção

ingredientes_sabores_norm_df = ingredientes_sabores_df.copy()

scaler = MinMaxScaler()

ingredientes_sabores_norm_df[todos_sabores] = scaler.fit_transform(ingredientes_sabores_norm_df[todos_sabores])
```

```python
ingredientes_sabores_norm_df[['ingrediente','categoria','sour', 'bitter', 'sweet', 'fruity', 'grassy','roast', 'caramel', 'fermented','powdery','dry']]
```

```python
# salvando a base tratada e normalizada em csv
ingredientes_sabores_norm_df.to_csv('./datasets/preprocessed/INGREDIENTES_normalizados_processed.csv', index=False)
```
